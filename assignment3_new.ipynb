## Calculate distance between two points in the earth(in terminal)

```python
### import libraries
import math
from math import radians,sin,cos,asin, sqrt
# get user inputs for calcualte distance
city1 = input("Please enter first city's name: ")
city2 = input ("Please enter second city's name: ")
lat1 = float(input("Please enter "+city1+"'s latitude: "))
lon1 = float(input("Please enter "+city1+"'s longitude: "))
lat2 = float(input("Please enter "+city2+"'s latitude: "))
lon2 = float(input("Please enter "+city2+"'s longitude: "))

ans = int(input("please select mesurment type\n 1 = Km, \n2 = Miles\n"))
if ans ==1:
    code = "Km"
    p=6377
else:
    code = "Miles"
    p = 3960
# Defining a fucntion to calculate distance    
def distance():
    #latitude and longitude to spherical coordinates in radians
    r_lat1 = lat1*math.pi/180
    r_lon1 = lon1*math.pi/180
    r_lat2 = lat2*math.pi/180
    r_lon2 = lon2*math.pi/180

    #calculate the angle φ from the North Pole down to the geographical location
    φ1 = (90-lat1)*math.pi/180
    φ2 = (90-lat2)*math.pi/180

    #degrees east of the prime meridian is the θ coordinate
    θ1 = lon1*math.pi/180
    θ2= lon2*math.pi/180

    x = sin(φ1) * sin(φ2)*cos(θ1-θ2)+cos(φ1)*cos(φ2)
    arc_length = math.acos(x)
    distance = arc_length*p
    print("Disttance from "+city1+" to "+city2+" is "+str(round(distance,2))+" "+code)
# call distance function
distance()
```
